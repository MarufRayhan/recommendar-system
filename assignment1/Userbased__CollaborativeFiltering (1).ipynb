To implement a user based collaborative filtering approach, at first we select a user. For example, here we choose an user having a userId of 100.

##Pearson Correlation Coefficient


```python
pearsonCorrelationDict = {}

for name, group in similarUsersGroup:
    
    group = group.sort_values(by='movieId')
    inputMovies = inputMovies.sort_values(by='movieId')
    
    temp_df = inputMovies[inputMovies['movieId'].isin(group['movieId'].tolist())]

    selectedUserTempRatingList = temp_df['rating'].tolist()

    similarUserTempRatingList = group['rating'].tolist()
    
    #Calculating pearson similarity
    similarityXX = 0
    meanSelectedUserTempRating = stat.mean(selectedUserTempRatingList)
    for i in selectedUserTempRatingList:
        similarityXX = similarityXX + pow((i - meanSelectedUserTempRating),2)
    
    similarityYY = 0
    meanSimilarUserTempRate = stat.mean(similarUserTempRatingList)
    for j in similarUserTempRatingList:
        similarityYY = similarityYY + pow((j - meanSimilarUserTempRate),2)
        
    similarityXY = 0
    for i, j in zip(selectedUserTempRatingList, similarUserTempRatingList):
        similarityXY = similarityXY+ ((i - meanSelectedUserTempRating ) * (j-meanSimilarUserTempRate))
    
  
    if similarityXX != 0 and similarityYY != 0:
        pearsonCorrelationDict[name] = similarityXY/np.sqrt(similarityXX*similarityYY)
    else:
        pearsonCorrelationDict[name] = 0      
```

```python
pearsonCorrelationDict.items()
```

```python
similarityScoreofSameTypeUsers_df = pd.DataFrame.from_dict(pearsonCorrelationDict, orient='index')
similarityScoreofSameTypeUsers_df.head()
```

```python
similarityScoreofSameTypeUsers_df.columns = ['similarityScore']
similarityScoreofSameTypeUsers_df['userId'] = similarityScoreofSameTypeUsers_df.index
similarityScoreofSimilarUsers_df.index = range(len(similarityScoreofSameTypeUsers_df))
similarityScoreofSimilarUsers_df.head()
```

### The top 10 similar users to input user

```python
topUsers=similarityScoreofSimilarUsers_df.sort_values(by='similarityScore', ascending=False)
topUsers.head(10)
```

## Rating of selected users to all movies

```python
topUsersRating = topUsers.merge(ratings, left_on='userId', right_on='userId', how='inner')
topUsersRating.head()
```

```python
# Calculating the mean ratings of all the similar users
meanRating = topUsersRating.groupby('userId').mean()[['rating']]
meanRating.columns = ['avgerageRating']
meanRating['userId'] = meanRating.index
meanRating.index = range(len(meanRating))
meanRating
```

```python
# Merging the top similar users average rating score 
topUsersRating = top=topUsersRating.merge(meanRating, left_on='userId', right_on='userId', how='inner')
topUsersRating.head()
```

```python
topUsersRating['weightedRatingScore'] = topUsersRating['similarityScore']*(topUsersRating['rating']-topUsersRating['avgerageRating'])
topUsersRating.head()
```

```python
#Applies a sum to the topUsers after grouping it up by userId
tempTopUsersRating = topUsersRating.groupby('movieId').sum()[['weightedRatingScore']]
tempTopUsersRating.columns = ['sum_weightedRatingScore']
tempTopUsersRating['movieId'] = tempTopUsersRating.index
tempTopUsersRating.head()
```

```python
recommendation_df= pd.DataFrame()

meanRa = inputMovies['rating'].mean()
recommendation_df['weighted average recommendation score'] = meanRa+(tempTopUsersRating['sum_weightedRatingScore']/topUsers['similarityScore'].sum())
recommendation_df['movieId'] = recommendation_df.index
recommendation_df.head()
```
